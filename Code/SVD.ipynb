```python
# DESCOPOSICION SVD 
import numpy as np

A = np.array([[0.6, 0.2], [0.4, 0.8]])

U, S, V = np.linalg.svd(A)

print(f"U: {U}")
print(f"S: {S}")
print(f"V: {V}")
```

```python
# OBTENER AUTOVALORES

A = np.array([[1, 2], [3, 4]])

autovalores, autovectores = np.linalg.eig(A)

print(f"Autovalores: {autovalores}")   
print(f"Autovectores: {autovectores}")
```

# COMPRESION DE IMAGENES

```python
from PIL import Image,ImageOps
import matplotlib.pyplot as plt
im = Image.open(r"dino.jpg")
imgris = ImageOps.grayscale(im)
```

```python
plt.imshow(imgris,cmap='gray')
```

```python
imagen_mat = np.array(list(imgris.getdata(band=0)), float)
imagen_mat.shape = (imgris.size[1], imgris.size[0])
print(imagen_mat.shape)
```

```python
U, S, VT = np.linalg.svd(imagen_mat)
```

```python
# Imagene reconstruida con 2 valores singulares
reconstructed_image = np.matrix(U[:,:2]) * np.diag(S[:2]) * np.matrix(VT[:2,:])
plt.imshow(reconstructed_image, cmap='gray')
plt.title('Reconstructed Image')
plt.show()
```

```python
# Imagene reconstruida con 60 valores singulares
reconstructed_image = np.matrix(U[:,:60]) * np.diag(S[:60]) * np.matrix(VT[:60,:])
plt.imshow(reconstructed_image, cmap='gray')
plt.title('Reconstructed Image')
plt.show()
```

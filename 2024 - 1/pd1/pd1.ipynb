```python
## PROBLEMA NUMERO 30
import math

def taylor_series_cos_approx(x, tolerance=1e-10):
    sum = 0
    term = 1  # Primer término de la serie de Taylor de cos(x) es 1
    n = 0  # Iniciamos con el término 0
    while abs(term) > tolerance:
        term = (-1)**n * x**(2*n) / math.factorial(2*n)
        sum += term
        n += 1
    return 1 - sum

# Prueba la función para varios valores de x
x_values = [-math.pi, -math.pi/2, 0, math.pi/2, math.pi]
for x in x_values:
    print(f"1 - cos({x}) = {taylor_series_cos_approx(x)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Nuestra función
def f(x):        
    return (-x + (x**2)/2 - (x**3)/3) / x
    
    
# Generar puntos x en el intervalo [-10^-15, 10^-15]
x_values = np.linspace(-1e-15, 1e-15, 20)
y_values = np.vectorize(f)(x_values)

# Graficar f(x) en el intervalo dado
plt.plot(x_values, y_values, label='f(x) usando $P_3(x)$')
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.title('Gráfica de $f(x) = \\frac{\\ln(1-x)}{x}$ cerca de $x = 0$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

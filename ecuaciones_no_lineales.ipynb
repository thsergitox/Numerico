# Método de la Bisección

```python
import numpy as np
import matplotlib.pyplot as plt

def biseccion(f, a, b, tol):
    fa = f(a)
    fb = f(b)
    i = 0
    a_list = [a]
    b_list = [b]
    c_list = []

    if f(a) * f(b) > 0:
        print("No hay raíz en el intervalo dado")
        return None

    while (b - a) > tol:
        i += 1
        c = (a + b) / 2
        a_list.append(a)
        b_list.append(b)
        c_list.append(c)
        
        print(f"El valor de c es: {c}")
        print(f"[a, b] = [{a}, {b}]")

        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c

    print(f"Numero de iteraciones: {i}")
    c_list.append((a + b) / 2)
    return (a + b) / 2, a_list, b_list, c_list

def f(x):
    return x**2 - 2

a = 1
b = 3
tol = 1e-5

root, a_list, b_list, c_list = biseccion(f, a, b, tol)
print(f"La raíz aproximada es: {root}")

# Graficar el proceso de bisección
x = np.linspace(a, b, 400)
y = f(x)

plt.plot(x, y, label="f(x)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(root, color='red', linestyle='--', label=f'Raíz aproximada: {root:.5f}')
plt.scatter(c_list, [f(ci) for ci in c_list], color='green', zorder=5, label='c en cada iteración')

for i in range(len(a_list)):
    plt.plot([a_list[i], b_list[i]], [0, 0], 'bo-')
    plt.plot([a_list[i], a_list[i]], [0, f(a_list[i])], 'b--', linewidth=0.5)
    plt.plot([b_list[i], b_list[i]], [0, f(b_list[i])], 'b--', linewidth=0.5)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Método de la Bisección')
plt.grid(True)
plt.show()
```

# Método de la regla falsa

```python
def regulafalsi(f, a, b):
    eps = 1e-3
    maxIter = 100
    fa = f(a)
    fb = f(b)
    i = 0
    
    while i < maxIter and b - a >= eps:
        c = a - fa * (b - a) / (fb - fa)
        fc = f(c)
        i += 1
        
        if fc == 0:
            a = b = c
        elif fa * fc < 0:
            b = c
            fb = fc
        else:
            a = c
            fa = fc
        
        print(f'[a, b] = [{a:.5f}, {b:.5f}]')
        print(f'Iteración {i}: c = {c:.5f}, f(c) = {fc:.5f}')
    
    if b - a >= eps:
        print('Metodo no converge')
    else:
        print(f'Solucion c = {c}')
        print(f'Numero de iteraciones = {i}')

# Función ejemplo
f = lambda x: x ** 2 - 2

# Parámetros de ejemplo
a = 1
b = 3

# Llamar a la función
regulafalsi(f, a, b)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 0 < alpha < 1
def regla_falsa_mod(f, a, b, tol, alpha=0.5):
    if f(a) * f(b) > 0:
        print("No hay raíz en el intervalo dado")
        return None

    # Inicializar valores de f(a) y f(b)
    fa = f(a)
    fb = f(b)
    
    c = a - fa * (b - a) / (fb - fa)
    fc = f(c)
    
    i = 0

    # Almacenar los valores para la gráfica
    a_list = [a]
    b_list = [b]
    c_list = [c]
    fa_list = [fa]
    fb_list = [fb]
    fc_list = [fc]

    while (b - a) > tol:
        i += 1
        print(f"El valor de c es: {c}")
        print(f"[a, b] = [{a}, {b}]")
        print(f"Iteración {i}")    
                
        if fc == 0:
            break
        elif fa * fc < 0:
            b = c
            fb = fc
            fa = alpha * fa  # Ajustar f(a) con alpha
        else:
            a = c
            fa = fc
            fb = alpha * fb  # Ajustar f(b) con alpha

        c = a - fa * (b - a) / (fb - fa)
        fc = f(c)
        
        # Almacenar los valores para la gráfica
        a_list.append(a)
        b_list.append(b)
        c_list.append(c)
        fa_list.append(fa)
        fb_list.append(fb)
        fc_list.append(fc)

    return c, a_list, b_list, c_list, fa_list, fb_list, fc_list

def f(x):
    return x**2 - 2

a = 1
b = 3
tol = 1e-5
root, a_list, b_list, c_list, fa_list, fb_list, fc_list = regla_falsa_mod(f, a, b, tol)
print(f"La raíz aproximada es: {root}")

# Graficar el proceso de Regula Falsi Modificado
x = np.linspace(a, b, 400)
y = f(x)

plt.plot(x, y, label="f(x)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(root, color='red', linestyle='--', label=f'Raíz aproximada: {root:.5f}')
plt.scatter(c_list, [f(ci) for ci in c_list], color='green', zorder=5, label='c en cada iteración')

for i in range(len(a_list) - 1):
    plt.plot([a_list[i], b_list[i]], [fa_list[i], fb_list[i]], 'bo-')
    plt.plot([a_list[i], a_list[i]], [0, fa_list[i]], 'b--', linewidth=0.5)
    plt.plot([b_list[i], b_list[i]], [0, fb_list[i]], 'b--', linewidth=0.5)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Método de Regula Falsi Modificado')
plt.grid(True)
plt.show()
```

# Método de Punto Fijo

```python
import numpy as np
import matplotlib.pyplot as plt

def g(x):
    return (x + 2/x) / 2  # Esta es la función g(x) = ∛(x + 1)

def punto_fijo(g, x0, tol, max_iter=100):
    x_list = [x0]
    i = 0

    while i < max_iter:
        i += 1
        x1 = g(x0)
        print(f"xk+1 = {x1}")
        print(f"Iteración {i}")
        x_list.append(x1)

        if abs(x1 - x0) < tol:
            break

        x0 = x1

    return x1, x_list

# Parámetros de ejemplo
x0 = 1.3
tol = 1e-5

root, x_list = punto_fijo(g, x0, tol)
print(f"La raíz aproximada es: {root}")

# Graficar el proceso del método del punto fijo
x = np.linspace(min(x_list) - 0.1, max(x_list) + 0.1, 400)
y = g(x)

plt.plot(x, x, 'r-', label='y = x')
plt.plot(x, y, label='g(x)')
plt.scatter(x_list, x_list, color='green', zorder=5, label='xk+1 en cada iteración')

for i in range(len(x_list) - 1):
    plt.plot([x_list[i], x_list[i]], [x_list[i], x_list[i+1]], 'b--', linewidth=0.5)
    plt.plot([x_list[i], x_list[i+1]], [x_list[i+1], x_list[i+1]], 'bo-')

plt.xlabel('x')
plt.ylabel('g(x)')
plt.legend()
plt.title('Método del Punto Fijo')
plt.grid(True)
plt.show()
```

# Metodo de la secante

```python
#Metodo de la secante

import numpy as np

def secante(f, x0, x1, tol):
    i = 0
    while True:
        i += 1
        x2 = x1 - f(x1)*(x1 - x0)/(f(x1) - f(x0))
        print(f"xk+1 = {x2}")
        print(f"Iteración {i}")          
        if abs(x2 - x1) < tol:
            return x2
        x0 = x1
        x1 = x2
        
def f(x):
    return x**2 - 2

x0 = 1
x1 = 3

tol = 1e-3

print(secante(f, x0, x1, tol))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def secante(f, x0, x1, tol):
    i = 0
    x0_list = [x0]
    x1_list = [x1]
    x2_list = []

    while True:
        i += 1
        x2 = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        print(f"xk+1 = {x2}")
        print(f"Iteración {i}")
        x2_list.append(x2)

        if abs(x2 - x1) < tol:
            break

        x0 = x1
        x1 = x2
        x0_list.append(x0)
        x1_list.append(x1)

    return x2, x0_list, x1_list, x2_list

def f(x):
    return x**2 - 2

x0 = 1
x1 = 3
tol = 1e-5

root, x0_list, x1_list, x2_list = secante(f, x0, x1, tol)
print(f"La raíz aproximada es: {root}")

# Graficar el proceso del método de la secante
x = np.linspace(min(x0_list + x1_list + x2_list), max(x0_list + x1_list + x2_list), 400)
y = f(x)

plt.plot(x, y, label="f(x)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(root, color='red', linestyle='--', label=f'Raíz aproximada: {root:.5f}')
plt.scatter(x2_list, [f(xi) for xi in x2_list], color='green', zorder=5, label='xk+1 en cada iteración')

for i in range(len(x0_list) - 1):
    plt.plot([x0_list[i], x1_list[i]], [f(x0_list[i]), f(x1_list[i])], 'bo-')
    plt.plot([x1_list[i], x2_list[i]], [f(x1_list[i]), 0], 'b--', linewidth=0.5)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Método de la Secante')
plt.grid(True)
plt.show()
```

# Método de Newton-Raphson

```python
# Metodo de Newton-Raphson

import numpy as np

def newton_raphson(f, df, x0, tol):
    i = 0
    while True:
        i += 1
        x1 = x0 - f(x0)/df(x0)
        print(f"xk+1 = {x1}")
        print(f"Iteración {i}")          
        if abs(x1 - x0) < tol:
            return x1
        x0 = x1
        
        
def f(x):
    try:
        return np.cos(2 * x)**2 - x**2
    except:
        return np.nan

def df(x):
    try:
        return -2 * np.sin(4 * x) - 2 * x
    except:
        return np.nan

x0 = 1.2

tol = 1e-10

print(newton_raphson(f, df, x0, tol))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def newton_raphson(f, df, x0, tol):
    i = 0
    x_list = [x0]

    while True:
        i += 1
        x1 = x0 - f(x0) / df(x0)
        print(f"xk+1 = {x1}")
        print(f"Iteración {i}")          
        x_list.append(x1)

        if abs(x1 - x0) < tol:
            break

        x0 = x1

    return x1, x_list

def f(x):
    return x**2 - 2

def df(x):
    return 2*x

x0 = 1
tol = 1e-5

root, x_list = newton_raphson(f, df, x0, tol)
print(f"La raíz aproximada es: {root}")

# Graficar el proceso del método de Newton-Raphson
x = np.linspace(min(x_list) - 0.1, max(x_list) + 0.1, 400)
y = f(x)

plt.plot(x, y, label="f(x)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(root, color='red', linestyle='--', label=f'Raíz aproximada: {root:.5f}')
plt.scatter(x_list, [f(xi) for xi in x_list], color='green', zorder=5, label='xk+1 en cada iteración')

for i in range(len(x_list) - 1):
    plt.plot([x_list[i], x_list[i+1]], [f(x_list[i]), 0], 'bo-')
    plt.plot([x_list[i+1], x_list[i+1]], [0, f(x_list[i+1])], 'b--', linewidth=0.5)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Método de Newton-Raphson')
plt.grid(True)
plt.show()
```

```python
import numpy as np

def f(x):
    try:
        return np.cos(2 * x)**2 - x**2
    except:
        return np.nan

def df(x):
    try:
        return -2 * np.sin(4 * x) - 2 * x
    except:
        return np.nan

def newton_raphson(f, df, x0, tol, max_iter=1000):
    for _ in range(max_iter):
        fx0 = f(x0)
        dfx0 = df(x0)
        if dfx0 == 0:
            return x0
        x1 = x0 - fx0 / dfx0
        if abs(x1 - x0) < tol:
            return x1
        x0 = x1
    return x1

# Parámetros
tol = 1e-10
interval = np.linspace(0, 1.5, 1500)

alpha = 0.5149
negative_alpha = -0.5149

# Inicializar listas para almacenar los subintervalos
converge_to_alpha = []
converge_to_minus_alpha = []

# Analizar la convergencia para cada valor inicial en el intervalo
for x0 in interval:
    root = newton_raphson(f, df, x0, tol)
    if np.isnan(root):
        continue
    if abs(root - alpha) < tol:
        converge_to_alpha.append(x0)
    elif abs(root - negative_alpha) < tol:
        converge_to_minus_alpha.append(x0)

# Identificar los subintervalos de convergencia
def find_intervals(x_list):
    intervals = []
    if not x_list:
        return intervals
    
    start = x_list[0]
    for i in range(1, len(x_list)):
        if x_list[i] - x_list[i-1] > 0.001:  # Ajustar el valor de diferencia según la precisión requerida
            intervals.append((start, x_list[i-1]))
            start = x_list[i]
    intervals.append((start, x_list[-1]))
    return intervals

alpha_intervals = find_intervals(converge_to_alpha)
minus_alpha_intervals = find_intervals(converge_to_minus_alpha)

print("Subintervalos que convergen a α ≈ 0.5149:")
for interval in alpha_intervals:
    print(f"[{interval[0]:.4f}, {interval[1]:.4f}]")

print("\nSubintervalos que convergen a -α ≈ -0.5149:")
for interval in minus_alpha_intervals:
    print(f"[{interval[0]:.4f}, {interval[1]:.4f}]")
```
